```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('weatherHistory.csv')
df.head()
```

```python
df.dtypes
```

```python
date=[]
for i in range(len(df)):
    date.append(df['Formatted Date'][i][:10]);
date
```

```python
df['Date']=date
df.head()
```

```python
new_df=df[['Date','Apparent Temperature (C)','Humidity']]
new_df
```

```python
new_df['Date']=pd.to_datetime(new_df['Date'])
```

```python
new_df['Date'][0]
```

```python
new_df.set_index('Date',inplace=True)
new_df
```

```python
new_df=new_df.resample('M').mean()
```

```python
new_df
```

```python
import matplotlib.pyplot as plt
```

# January

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.plot(new_df[new_df.index.month==1]['Humidity'])
plt.title('Humidity in January from 2006-2016')
plt.subplot(1,2,2)
plt.plot(new_df[new_df.index.month==1]['Apparent Temperature (C)'],color='red')
plt.title('Apparent Temperature in January from 2006-2016')
plt.show()
```
